```python
def dist(x,y):
    dis=0.0
    for i in range(0,22):
        dis+=(x[i]-y[i])*(x[i]-y[i])
    return dis
```

```python
def KmeansClustering():
    best_acc = 0.0
    c1 = [0 for i in range(0, 22)]
    c2 = [0 for i in range(0, 22)]
    best_c1 = [0 for i in range(0, 22)]
    best_c2 = [0 for i in range(0, 22)]
    best_c1_data_points=[]
    best_c2_data_points=[]
    t=10
    while t > 0:
        c1=dataset[random.randint(0,194)]
        c2=dataset[random.randint(0,194)]
        #print c1
        #print c2
        c1data=[0 for i in range(0, 22)]
        c2data=[0 for i in range(0, 22)]
        c1_count=0
        c2_count=0
        c1_data_points=[]
        c2_data_points=[]
        for j in xrange(0,len(dataset)):
            sample=dataset[j]
            d1=dist(sample,c1)
            d2=dist(sample,c2)
            if d1<d2:
                c1_count+=1
                for i in range(0,22):
                    c1data[i]+=sample[i]
                c1_data_points.append(j)   
            else:
                c2_count+=1
                for i in range(0,22):
                    c2data[i]+=sample[i]
                c2_data_points.append(j)    
        #print c1_count
        #print c2_count
        for i in range(0,22):
            c1[i]=(c1data[i]*1.0)/c1_count
        for i in range(0,22):
            c2[i]=(c2data[i]*1.0)/c2_count
        #print c1
        #print c2
        #print t
        #assume c1 to be 1 cluster
        correct=0
        total=195
        i=0
        for sample in dataset:
            d1=dist(sample,c1)
            d2=dist(sample,c2)
            #print d1
            #print d2

            if d1 < d2:
                if labels[i] == "1":
                    correct+=1
            else:
                if labels[i] == "0":
                    correct+=1
            i+=1
        #print correct
        acc = correct/195.0
        if acc > best_acc:
            best_acc = acc
            best_c1_data_points=c1_data_points[:]
            best_c2_data_points=c2_data_points[:]
            for j in range(0, 22):
                best_c1[j] = c1[j]
                best_c2[j] = c2[j]
        t=t-1
    best_c1 = np.array(best_c1)
    best_c2 = np.array(best_c2)
    
    means_1=[]
    means_2=[]
    for i in xrange(0,22):
        means_1.append(0)
        means_2.append(0)
    
    for datapoint in best_c1_data_points:
        for j in xrange(0,22):
            means_1[j]+=dataset[datapoint][j]
            
    for datapoint in best_c2_data_points:
        for j in xrange(0,22):
            means_2[j]+=dataset[datapoint][j]
            
    for j in xrange(0,22):
        means_1[j]=(1.0*means_1[j])/len(best_c1_data_points)
    
    for j in xrange(0,22):
        means_2[j]=(1.0*means_2[j])/len(best_c2_data_points)
    
    for i in range(0, 22):
        ratio_c1[i] = (means_1[i]*1.0)/best_c1[i]
        ratio_c2[i] = (means_2[i]*1.0)/best_c2[i]
    #np.array(ratio_c1)
    #np.array(ratio_c2)
    
    c1_rat = np.array(ratio_c1)
    c2_rat = np.array(ratio_c2)
    #print c1_rat
    #print c2_rat
    i=0
    for x in range(0, 195):
        d1=dist(sample,c1)
        d2=dist(sample,c2)
        if d1 < d2:
            if labels[i] == "1":
                for i in range(0, 22):
                    dataset[x][i] *= c1_rat[i]
        else:
            if labels[i] == "0":
                for i in range(0, 22):
                    dataset[x][i] *= c2_rat[i]
        i+=1
```

```python
class ComplexNumber():
    def __init__(self,real,complx):
        self.real=real
        self.complx=complx
    def getReal(self):
        return self.real
    def getComplex(self):
        return self.complx
```

```python
def multiply(z1,z2):
    return ComplexNumber(((z1.getReal())*(z2.getReal()))-((z1.getComplex())*(z2.getComplex())),(((z1.getReal())*(z2.getComplex()))+((z1.getComplex())*(z2.getReal()))))
```

```python
def add(z1,z2):
    return ComplexNumber(z1.getReal()+z2.getReal(),z1.getComplex()+z2.getComplex())
```

```python
def subtract(z1,z2):
    return ComplexNumber(z1.getReal()-z2.getReal(),z1.getComplex()-z2.getComplex())
```

```python
def MatrixMultiply(matrixA,matrixB):
    rowA=len(matrixA)
    colA=len(matrixA[0])
    rowB=len(matrixB)
    colB=len(matrixB[0])
    answer=[]
    
    if(colA!=rowB):
        print "Matrix Multiply Error"
        return answer
    
    for i in xrange(0,rowA):
        tempArray=[]
        for j in xrange(0,colB):
            temp=ComplexNumber(0,0)
            for k in xrange(0,colA):
                temp=add(temp,multiply(matrixA[i][k],matrixB[k][j]))
            tempArray.append(temp)
        answer.append(tempArray)
    return answer
```

```python
def sigmoid(number):
    return expit(number)
```

```python
def convertToOutput(matrixA):
    rowA=len(matrixA)
    for i in xrange(0,rowA):
        temp=[]
        temp.append(ComplexNumber(sigmoid(matrixA[i][0].getReal()),sigmoid(matrixA[i][0].getComplex())))
        matrixA[i]=temp[:]
    return matrixA
```

```python
def transpose(matrixA):
    rowA=len(matrixA)
    colA=len(matrixA[0])
    matrixATranspose=[]
    for j in xrange(0,colA):
        tempArray=[]
        for i in xrange(0,rowA):
            tempArray.append(matrixA[i][j])
        matrixATranspose.append(tempArray)
    return matrixATranspose
```

```python
def elementWiseMultiplication(matrixA,matrixB):
    rowA=len(matrixA)
    colA=len(matrixA[0])
    rowB=len(matrixB)
    colB=len(matrixB[0])
    
    answerMatrix=[]
    if rowA!=rowB:
        print "Element Wise Multiplication Error"
        return answerMatrix
    
    for i in xrange(0,rowA):
        tempArray=[]
        for j in xrange(0,colA):
            tempArray.append(multiply(matrixA[i][j],matrixB[i][j]))
        answerMatrix.append(tempArray)
    return answerMatrix
```

```python
def convertToMatrix(z):
    temp=[]
    tempArray=[]
    tempArray.append(z)
    temp.append(tempArray)
    return temp
```

```python
def elementWiseMultipliedMatrix(matrixA):
    elementWiseMultiplied=[]
    for i in xrange(0,len(matrixA)):
            tempArray=[]
            tempArray.append(multiply(matrixA[i][0],subtract(ComplexNumber(1,1),matrixA[i][0])))
            elementWiseMultiplied.append(tempArray)
    return elementWiseMultiplied
```
